```python
# import libraries
import numpy as np
import pandas as pd
```

```python
titanic_train = pd.read_csv("titanic_train.csv")
titanic_test = pd.read_csv("titanic_test.csv")
```

```python
titanic_train.head()
```

```python
titanic_test.head()
```

```python
titanic_train.info()
```

```python
titanic_test.info()
```

```python
titanic_test["Survived"] = -10
```

```python
titanic_test.columns
```

```python
titanic_combined = pd.concat((titanic_train,titanic_test),axis=0)
titanic_combined.head()
```

```python
titanic_combined.info()
```

```python
titanic_combined = titanic_combined.reset_index()
```

```python
titanic_combined.drop("index",axis = 1,inplace=True)
```

```python
titanic_combined.head()
```

```python
# check na values
titanic_combined.isna().sum()
```

```python
titanic_combined['Age'].dtype
```

# If I use fillna method , the dtype of Age is getting changed to object so used transform method instead

```python
# get the mean of Age group by sex
x = titanic_combined.groupby('Sex').Age.transform('median')
titanic_combined["Age"] = np.where(titanic_combined.Age.isnull(), x, titanic_combined.Age)
```

```python
# fill na values in cabin with NC
titanic_combined['Cabin'].fillna('NC',inplace=True)
```

```python
#mean fare based on sex
mean_fare = titanic_combined.groupby('Sex').Fare.transform('mean')
titanic_combined["Fare"] = np.where(titanic_combined.Fare.isnull(),mean_fare, titanic_combined.Fare)
```

```python
titanic_combined.isna().sum()
```

```python
titanic_combined[titanic_combined['Embarked'].isna()==True]
```

```python
titanic_combined.dropna(inplace=True)
```

```python
titanic_combined.info()
```

```python
titanic_combined.set_index('PassengerId',inplace=True)
```

```python
titanic_combined.head()
```

```python
titanic_combined['family_size'] = titanic_combined['Parch'] + titanic_combined['SibSp']
```

```python
titanic_features = titanic_combined[['Age','Embarked','Fare','Name','Pclass','Sex','family_size','Survived']].copy()
titanic_features.head()
```

```python
# create age group column 
titanic_features["Agegroup"] = pd.cut(titanic_features.Age, bins = [0, 2, 5, 18,60,100], labels=["Infant","Toddler","Child","Adult","Senior"],include_lowest=True)
```

```python
titanic_features.drop('Age',axis=1,inplace=True)
titanic_features.head()
```

```python
# create fare group column 0 ,1 ,2, 3 - very_low , low , high , very_high
titanic_features['Fare_group'] = pd.qcut(titanic_features.Fare, 4, labels=[0,1,2,3])
titanic_features.head()
```

```python
titanic_features.drop('Fare',axis=1,inplace=True)
titanic_features.head()
```

```python
# extract title from name
titanic_features["Title"] = titanic_features.Name.str.split("[,.]").str.get(1).str.lower().str.strip()
```

```python
titanic_features.head()
```

```python
def getTitle(title):
    title_group = {'mr' : 'Mr', 
               'mrs' : 'Mrs', 
               'miss' : 'Miss', 
               'master' : 'Master',
               'don' : 'Sir',
               'rev' : 'Sir',
               'dr' : 'Officer',
               'mme' : 'Mrs',
               'ms' : 'Mrs',
               'major' : 'Officer',
               'lady' : 'Lady',
               'sir' : 'Sir',
               'mlle' : 'Miss',
               'col' : 'Officer',
               'capt' : 'Officer',
               'the countess' : 'Lady',
               'jonkheer' : 'Sir',
               'dona' : 'Lady'
                 }
    return title_group[title]
titanic_features['Title'] =  titanic_features.Title.map(lambda x : getTitle(x))
```

```python
titanic_features['Title'].value_counts()
```

```python
titanic_features.drop('Name',axis=1,inplace=True)
titanic_features.head()
```

```python
titanic_features.shape
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
titanic_features['Title'] = number.fit_transform(titanic_features['Title'].astype('str'))
```

```python
titanic_features['Embarked'] = number.fit_transform(titanic_features['Embarked'].astype('str'))
```

```python
titanic_features['Sex'] = number.fit_transform(titanic_features['Sex'].astype('str'))
```

```python
titanic_features = titanic_features.rename({"Sex":"isMale"}, axis='columns')
```

```python
titanic_features.head()
```

```python
titanic_features = pd.get_dummies(titanic_features,prefix=['A'],columns=['Agegroup'],drop_first=True)
titanic_features.head()
```

# split back to train and test sets

```python
titanic_test = titanic_features[titanic_features['Survived']==-10]
titanic_test.head()
```

```python
titanic_test.shape
```

```python
titanic_train = titanic_features[titanic_features['Survived']!=-10]
titanic_train.head()
```

```python
titanic_train.shape
```

```python
x_train = titanic_train[['Embarked','Pclass','isMale','family_size','Fare_group','Title','A_Toddler','A_Child','A_Adult',"A_Senior"]].copy()
y_train = titanic_train[['Survived']].copy()
x_test = titanic_test[['Embarked','Pclass','isMale','family_size','Fare_group','Title','A_Toddler','A_Child','A_Adult',"A_Senior"]].copy()
y_test = titanic_test[['Survived']].copy()
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train.dtypes
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression(solver='liblinear',random_state=10)
```

```python
Lr.fit(x_train,y_train)
```

```python
Lr.predict(x_test)
```

```python
# predict on train set and get the accuracy , since we don't have actual values for test set to predict the accuracy
y_pred = Lr.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_train,y_pred)
accuracy
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
knn.predict(x_test)
```

```python
y_pred_knn = knn.predict(x_train)
accuracy_knn = accuracy_score(y_train,y_pred_knn)
accuracy_knn
```

# Naive-bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.predict(x_test)
```

```python
y_pred_gnb = gnb.predict(x_train)
accuracy_gnb = accuracy_score(y_train,y_pred_gnb)
accuracy_gnb
```
